## Preparing Data  

```python
import twder
import pandas as pd
from datetime import datetime

# in the same trading day, p6m and p1d are identical
def FE_form(df):
    df.columns = ['date', 'cashbuy', 'cashsell', 'buy', 'sell']
    df['date'] = pd.to_datetime(df['date'])
    
    return pd.concat([df['date'], df['sell']], 1)

def past6mon(currency):
    temp = pd.DataFrame(twder.past_six_month(currency))
    df = FE_form(temp)
    
    return df

def past1day(currency):
    temp = pd.DataFrame(twder.past_day(currency))
    df = FE_form(temp)
    
    return df
```
